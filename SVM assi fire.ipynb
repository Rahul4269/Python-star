```python
df=df.drop(columns=['month', 'day'], axis=0)
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
new_data=scaler.fit_transform(df.drop(labels=['size_category'],axis=1))
```

```python
columns=['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind',
       'rain', 'area', 'dayfri', 'daymon', 'daysat', 'daysun', 'daythu',
       'daytue', 'daywed', 'monthapr', 'monthaug', 'monthdec', 'monthfeb',
       'monthjan', 'monthjul', 'monthjun', 'monthmar', 'monthmay', 'monthnov',
       'monthoct', 'monthsep']
```

```python
new_df=pd.DataFrame(data=new_data,columns=columns)
```

```python
x=new_df
y=df['size_category']
```

```python
from  sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
model.fit(train_x,train_y)
```

```python
model.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test_y,model.predict(test_x))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.1,1,10,50,100,500],'gamma':[1,0.5,0.1,0.01,0.001]}
```

```python
grid= GridSearchCV(SVC(),param_grid, verbose=3, n_jobs=-1)
```

```python
grid.fit(train_x,train_y)
```

```python
grid.best_params_
```

```python
model_new=SVC(C=500, gamma=0.01)
model_new.fit(train_x,train_y)
```

```python
accuracy_score(test_y,model_new.predict(test_x))
```
